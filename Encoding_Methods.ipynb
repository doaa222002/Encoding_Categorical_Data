# Handling Catigorical Data By Encoding


# Import Libraries 

```python
import numpy as np 
import pandas as pd 
import category_encoders as ce 
import sklearn
```

```python
# !pip install category_encoders
```

```python
# creating dataset 
data= pd.DataFrame ({
    'gender': ['male ','female','male','female','female'],
    'class' :['A','B','C','D','A'],
    'city' :['cairo','giza','cairo','cairo','giza']
})
data
```

# 1- One Hot Encoder

```python
#call of one hot encoder
one_hot_encoder=ce.OneHotEncoder(cols=['gender','city'])

#use fit_transform one hot encoder on the data 
new_data=one_hot_encoder.fit_transform(data)
new_data
```

```python
new_data=new_data.rename(columns={'gender_1':'male','gender_2':'female','city_1':'cairo','city_2':'giza' })
new_data
```

## 2-Lable Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.DataFrame({
    'Gender':['male','female ','male','female '],
    'school':['sool','sec','third','fourth'],
    'courses':['AI','WEB','JAVA','FLUTTTER'],
    'city':['cairo ', 'giza','cairo ', 'giza']
})
data
```

```python
#call LabelEncoder 
label_encoder=LabelEncoder()
#fit transform label_encoder on 'school' column 
encoded_data=label_encoder.fit_transform(data['school'])
```

```python
encoded_data
```

```python
encoded_data=pd.DataFrame(encoded_data)
```

```python
encoded_data=encoded_data.rename(columns={0:'Encoded_school'})
encoded_data=pd.concat([data,encoded_data],axis=1)
```

```python
encoded_data
```

```python
encoded_data.drop(columns="school",inplace=True)
encoded_data
```

```python
#call label encoder from sklearn
label_encoder=LabelEncoder()
#fit transform label_encoder on 'courses' column
encoded_courses=label_encoder.fit_transform(data['courses'])
encoded_courses
```

```python
encoded_courses=pd.DataFrame(encoded_courses)
```

```python
encoded_courses=encoded_courses.rename(columns={0:'Encoded_courses'})
encoded_courses
```

```python
encoded_data=pd.concat([encoded_courses,encoded_data],axis=1)
encoded_data
```

## 3-Binary Encoder

```python
new_data['class'].unique()
```

```python
# call binary encoder 
binary_enconder=ce.BinaryEncoder(cols=['class'])
#fit and transform data with binary encoder
new_data =binary_enconder.fit_transform(new_data)
new_data
```

## 4- Ordinal Encoder

```python
data=pd.DataFrame({
    "height":['medium','short','tall','medium','short','tall','medium','short','tall']
})
data
```

```python
ordinal_encoding=ce.OrdinalEncoder(cols = ['height'],
                                  return_df=True,
                                  mapping=[{'col':'height',
                                           'mapping':{
                                               'none':0,
                                               'tall':1,
                                               'medium':2,
                                               'short':3
                                           }}])
```

```python
data['encoding']=ordinal_encoding.fit_transform(data)
data
```

```python
data.drop('height',axis=1,inplace=True)
data
```

## 5-Target Encoding

```python
df=pd.DataFrame({
    'name':['alex','john','marry','alex','john','marry'],
    'marks':[10,20,30,40,50,60]
})
df
```

```python
#apply target encoding 
target_encoder=ce.TargetEncoder(cols='name')
#fit transform target encoding 
target_encoder.fit_transform(df['name'],df['marks'])
```

## 6- Hash Encoder

```python
data=pd.DataFrame({
    "color":['blue ','green','blue','black','yellow','black','green']
})
data
```

```python
data['color'].unique()
```

```python
hash_encoder=ce.HashingEncoder(cols=['color'],n_components=5)
hash_encoder.fit_transform(data)
```
